```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/penguins_binary_classification.csv")
df
```

```python
df.describe(include="all")
```

```python
df['island'] = df['island'].map({'Torgersen':0,'Biscoe':1,'Dream':2})
```

```python
df.head()
```

```python
df.describe(include="all")
```

```python
df.drop(['year','island'],axis=1,inplace=True)
```

```python
X = df.drop('species',axis=1)
Y = df['species']
```

```python
X_train , x_test , Y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
log_model = LogisticRegressionCV(Cs=[1,10,50,100,150,1000],max_iter=1000)
```

```python
log_model.fit(X_train,Y_train)
```

```python
log_model.coef_
```

```python
log_model.C_
```

```python
y_pred = log_model.predict(x_test)
```

```python
y_pred_proba = log_model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_pred))
```
